# Importing Required Libraries

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# reading the image
pic = cv2.imread("image.png")

# Plotting the histogram 
plt.hist(pic.ravel(), 255, [0, 255])
plt.show()
```

```python
# Separate Colour Channels and plotting them all in 1 histogtram!
# TO check the description of a function press shift + tab after placing the mouse pointer on the function.

color = ["b", "g", "r"]

for i, col in enumerate(color):
    his = cv2.calcHist([pic], [i], None, [256], [0, 256])
    plt.plot(his, color = col)
    plt.xlim([0, 256])
plt.show()    
```
